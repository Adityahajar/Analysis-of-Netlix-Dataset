```python
plt.title("Date of Last Payment ")
sns.histplot(df, x="Last Payment Date", bins=10, kde=True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.title("Age")
sns.histplot(df, x="Age", bins=20, kde=True)
plt.ylabel("Count of Age")
plt.show()
```

* Maximum Susbcription are from the age group of 26-30
* Netflix Has More Users and Subscribers are from the age group of 26-30.

```python
plt.title("Subscriptions From Male and Female Users ")
sns.countplot(df, x="Gender", hue="Subscription Type")
plt.show()
```

* Subscriptions From the male and female users are aproximetly equal
* Somewhere in basic type of Subscription Female user are more than the Male User. 

```python
plt.title("Subscription ON  Device")
sns.countplot(df, x="Device", hue="Device")
plt.show()
```

```python
df["Age Group"] = pd.cut(df["Age"], bins=5, 
                         labels=["26-30", "31-35", "36-40", "41-45", "45-More"])
```

```python
df.head()
```

```python
pd.crosstab(columns=df["Subscription Type"], index=df["Age Group"])
```

```python
sns.heatmap(pd.crosstab(columns=df["Subscription Type"],
                        index=df["Age Group"]), annot=True, fmt="d", cmap="viridis")
plt.show()
```

##### Insights : By using this we can clearly see that the basic subscription is more often subscribed by 26-30 age group In 41-45 age group the subscription is less compared to other and that is Premium type

```python
sns.heatmap(pd.crosstab(columns=df["Subscription Type"],
                        index=df["Device"]), annot=True, fmt="d", cmap="viridis")
plt.show()
```

##### Insights: By Using That we Can Clearly Undrstand The in Laptop  Device Users Has the More Basic Type Of   Subscriptions as wll as Premimum Type of Subscription.
